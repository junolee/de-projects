```python
import boto3
import yaml

# FILE_NAME = 'customer_full_data.csv'
FILE_NAME = 'customer_change_data.csv'

BUCKET_NAME = 'jl-scd1-data-bucket'
FOLDER_NAME = 'data'

with open('aws_credentials.yml', 'r') as file:
	aws_credentials = yaml.safe_load(file)

# Initialize session using AWS credentials from YAML
session = boto3.Session(
	aws_access_key_id=aws_credentials['aws_access_key_id'],
	aws_secret_access_key=aws_credentials['aws_secret_access_key'],
	region_name=aws_credentials['aws_region_name']
)

# Create an S3 client
s3_client = session.client('s3')

try:
# Upload file in S3
	s3_client.upload_file(
		Filename=FILE_NAME,
		Bucket=BUCKET_NAME,
		Key=FOLDER_NAME + '/' + FILE_NAME
	)
	print(FILE_NAME + " is uploaded sucessfully")

except Exception as e:
	print(f"An error occurred: {e}")
```
